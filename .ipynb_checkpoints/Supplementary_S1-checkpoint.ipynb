```python
import numpy, pybiomart, pandas, seaborn

import scanpy
scanpy.set_figure_params(dpi_save=1000)
```

```python
import matplotlib, matplotlib.pyplot
matplotlib.rcParams.update({'font.size':20, 
                            'font.family':'FreeSans', 
                            'xtick.labelsize':20, 
                            'ytick.labelsize':20, 
                            'figure.figsize':(12, 8)})
```

# input data

```python
iso1 = 'ENST00000354449.7'
iso2 = 'ENST00000354956.9'
```

```python
%%time
input_file = '/home/adrian/databases/gtex/zscore.tsv'
adata = scanpy.read_csv(input_file, delimiter='\t')
adata
```

```python
%%time

neis = [3, 4, 5, 6, 7, 8, 10, 20, 30, 50, 100, 200, 500, 1000]

cooccs = []; ks = []
dist = 0.2
spre = 10

for nei in neis:
    scanpy.pp.neighbors(adata, n_neighbors=nei, n_pcs=100)
    scanpy.tl.leiden(adata)
    scanpy.tl.paga(adata)
    scanpy.pl.paga(adata, plot=False)
    scanpy.tl.umap(adata, min_dist=dist, spread=spre, init_pos='paga')  
    
    iso1_cluster = None; iso2_cluster = None
    for category in range(len(adata.obs['leiden'].cat.categories)):
        if str(category) == adata.obs['leiden'][iso1]:
            iso1_cluster = str(category)
        if str(category) == adata.obs['leiden'][iso2]:
            iso2_cluster = str(category)
    
    # check cooccurrence
    #print(iso1_cluster, iso2_cluster)
    if iso1_cluster == iso2_cluster:
        coocc = 1
    else:
        coocc = 0
    cooccs.append(coocc)
    
    # number of clusters
    number_of_clusters = len(adata.obs['leiden'].cat.categories)
    ks.append(number_of_clusters)
```

```python
print(neis)
print(numpy.min(cooccs), cooccs)
print(ks)
```

```python
for i in range(len(neis)):
    if cooccs[i] == 1:
        matplotlib.pyplot.plot(neis[i], ks[i], 'o', color='tab:red', markersize=18)
    else:
        matplotlib.pyplot.plot(neis[i], ks[i], 'o', color='tab:green', markersize=18)

matplotlib.pyplot.plot(neis, ks, ls='--', color='black', alpha=2/3, zorder=0)

legend_elements = [ 
    matplotlib.lines.Line2D([0], [0], color='tab:red', marker='o', markerfacecolor='tab:red', markeredgecolor='white', markersize=18),
    matplotlib.lines.Line2D([0], [0], color='tab:green', marker='o', markerfacecolor='tab:green', markeredgecolor='white', markersize=18),
     ]
matplotlib.pyplot.legend(legend_elements, ['Isoforms co-occurrence', 'No co-occurrence'], ncol=1, loc=1)

matplotlib.pyplot.xlabel('Louvain number of neighbours')
matplotlib.pyplot.ylabel('Obtained number of clusters')

matplotlib.pyplot.grid(alpha=0.5, ls='--')
matplotlib.pyplot.tight_layout()

#matplotlib.pyplot.show()
matplotlib.pyplot.savefig('occurence.svg')
```
